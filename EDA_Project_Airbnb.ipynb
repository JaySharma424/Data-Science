This capstone project aims to explore throughan Airbnb bookings dataset, the preference of customers, the behavior of hosts, pricing patterns and availability trends. As a global hospitality platform, Airbnb has petabytes of data collected from millions of listings and users. This data analysis could reveal useful insights to restore confidence-based decision making as it relates to hosts, customers and business stakeholders.

The.csv file loaded has near 49,000 observations and 16 attributes, mixing both categorical and numerical variables. Listing details provided include room_type, price, neighbourhood_group, minimum_nights, number_of_reviews, reviews_per_month and availability_365. These factors give a complete depiction of the listing properties, location-centric demand and user interactions.

Data understanding and cleaning: The analysis was started by understanding the data, removing values that are missing, incorrect data types, or outliers. There were missing values in columns like last_review and reviews_per_month, which were imputed properly so as not to introduce any bias into the analysis. We cleaned the price variable to avoid unrealistic outliers that might mislead our visualisation.

After cleaning we perfomed Univariate and Bivariate analysis with additional inforamtion using python libraries: Pandas, NumPy, Matplotlib and Seaborn. Several plots were generated to meet the project's requirement to include at least 5 LDAs. These were bar charts, box plots, histograms, scatter plots and heatmaps.

An interesting result from the study was in room type choice. Entire home/apt and private room are the most popular listings type, private rooms are lower-priced compare to entire homes/apartments but still relatively in demand for budget travelers. Whole homes—even with the higher price tag—also generate more engagement in premium markets.

There was wide variability in pricing analysis between neighbourhood groups. They also share higher average prices in the central and most well-trafficked neighborhoods, suggesting strong demand. Box plots showed most of the listings are in a reasonable price range, except fine few luxury listings that drive up the maximum price.

There was an inverse relationship between the price and number of reviews for a listing, meaning that cheaper listings are reviewed more - What happens next predicts this hypothesised correlation trend. Moreover, those listings which have highest availability_365 all tended to have less reviews, that is to say those high-availability-turnover apartments may be not so popular or seasonal sesame.

Minimum nights is also something that came with a huge revelation for us. Properties with higher minimum night requirement receive less customer engagement suggesting that shorter stays are preferred by majority of users on this platform.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code

# Select numerical columns
num_df = df.select_dtypes(include=np.number)

# Correlation matrix
corr = num_df.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used to understand the strength and direction of relationships between numerical variables. It helps identify which variables are positively or negatively related and whether any features are redundant.

##### 2. What is/are the insight(s) found from the chart?

1) number_of_reviews and reviews_per_month show a strong positive correlation, indicating that frequently booked listings receive more monthly reviews.

2) price has a weak or negative correlation with number_of_reviews, suggesting higher-priced listings tend to receive fewer bookings.

3) availability_365 shows a negative correlation with reviews, indicating popular listings are booked more often and thus less available.

#### Chart - 15 - Pair Plot

##### 1. Why did you pick the specific chart?

A pair plot helps visualize pairwise relationships between multiple numerical variables simultaneously. It is useful for detecting trends, clusters, and potential linear relationships.

Clear patterns between price and number_of_reviews confirm an inverse relationship.

availability_365 and reviews_per_month show opposite trends, reinforcing demand-related insights.
