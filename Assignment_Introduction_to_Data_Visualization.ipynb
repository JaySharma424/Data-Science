```
total_profit = df['total_profit']
total_month = df['month_number']
plt.bar(total_month,total_profit)
plt.title('Total Profit Accross the month')
plt.xlabel('Months')
plt.ylabel('Total Profit')
print(total_profit)
```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
df.plot(x='month_number',y=['facecream', 'facewash'],kind='bar')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales Data')
plt.show()
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here

df.plot(x = 'month_number', y = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'], kind = 'line')
plt.title('Units Sold per month using Multiline Plots')
plt.xlabel('number of months')
plt.ylabel('Product Sales')
```
